```python
#4. Opening CSV File from an URL

import pandas as pd
import requests
import io

url = "https://raw.githubusercontent.com/Tanishqa-10/AskPython/main/Sampledata.csv"
response = requests.get(url)
s = response.content

# Convert bytes to a file-like object
csv_file = io.BytesIO(s)

# Read the CSV file using pandas
c = pd.read_csv(csv_file)

# Now you can use 'c' which should contain your CSV data
```

```python
#5. Sep Parameter

#using sep parameter for data which are seperated into tab so it will be show using sep function

pd.read_csv('movie.metadata.tsv', sep = '\t', names = ['Id Num' , 'Employee Code', 'Name ','DOB' ])
```

```python
#6. Index_Col Parameter
#Any collum which was not usefull in Data Analytis or Machine Learning
#We can convert those collum into index
#Default padnasd index will be replace with enrolle id index


pd.read_csv('aug_train.csv', index_col ='enrollee_id' )
```

```python
#7. Header Parameter
#first row will be start form 1 and 0th colloum will be deleted
#one collumn from top will be deleted

pd.read_csv('aug_train.csv',header = 1)
```

```python
#8. Use_Collumn Parameter
#When you are importing the data you can specify particular collumn of those data set which you will be needed and only those collum will be extracted form the data


pd.read_csv('aug_train.csv', usecols=['enrollee_id', 'city', 'gender','education_level'])
```

```python
#9. Squeeze Parameters
#When you need only one collumn form data set and you use squeeze it will so it will become pandas series


pd.read_csv('aug_train.csv', usecols= ['enrollee_id'], squeeze = True)
```

```python
#10. Skiprows/nrows Parameter
#A particlar rows will be skip form you data set while you are importing the data set
#top 2 rows has been skipped

pd.read_csv('aug_train.csv', skiprows=[0,1])
```

```python
#Nrows Parameter
#It will give only limited number of rows
#Index form 0 to 99 will  be printed

pd.read_csv('aug_train.csv',nrows = 100)
```

```python
#10. Econding Parameter
#a common error that occurs when trying to read a file with Pandas that contains non-UTF-8 encoded characters


pd.read_csv('zomato.csv', encoding= 'latin-1')
```

```python
#11. Skip Bad Lines

pd.read_csv('College_Data.csv')
```

```python
# Skip Bad Lines
#All the empty collumns will be converted into '''' empty collumn

pd.read_csv('College_Data.csv', sep=';', encoding='latin-1', error_bad_lines=False)
```

```python
#12. Dtypes Parameter
#Dtype have many folat values so dtype function will converted float values into int
pd.read_csv('aug_train.csv').info()
```

```python
#13. Handling Dates
#When you are importing any data then the date will be treated as string
#date will converted into objects (objects = String in pandas)

pd.read_csv('matches.csv', parse_dates = ['date'])

#object will conveted into datetime data type
#and we can put filter in particular dates
```

```python
#14. Name Changes

def rename(name):
  if name == "Sunrisers Hyderabad":

    return "SRH"
  else:
    return name
```

```python
rename("Sunrisers Hyderabad")
```

```python

pd.read_csv('matches.csv')
```

```python
#Converted

pd.read_csv('matches.csv', converters={'team1' : rename})
```

```python
#15. Na_Values Parameter
#In we can take example as male and female if we select male then all male collumn will be Nan

pd.read_csv('College_Data.csv', na_values = ['male'])
```
